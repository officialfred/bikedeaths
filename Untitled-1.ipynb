```python
url_alt =  "https://data.cityofnewyork.us/resource/h9gi-nx95.geojson?$where=%20number_of_cyclist_killed%20%3E0&$limit=10000"
url = "https://data.cityofnewyork.us/resource/h9gi-nx95.geojson?$where=%20number_of_persons_killed%20%3E0%20AND%20crash_date%20%3E%20%222020-01-01%22%20&$limit=10000"

gdf = gpd.read_file(url)
gdf.head()
```

```python
gdf.columns
```

```python
gdf = gdf[[ 'crash_date', 'crash_time',
         'on_street_name', 'off_street_name', 
       'number_of_cyclist_killed', 'number_of_motorist_killed', 'number_of_pedestrians_killed',
       'vehicle_type_code1', 'contributing_factor_vehicle_1',
       'vehicle_type_code2', 'contributing_factor_vehicle_2',
       'vehicle_type_code_3','contributing_factor_vehicle_3',
       'vehicle_type_code_4', 'contributing_factor_vehicle_4',
       'vehicle_type_code_5', 'contributing_factor_vehicle_5', 
       'geometry']]

gdf.rename(columns={'vehicle_type_code1': 'vehicle_type_code_1', 'vehicle_type_code2': 'vehicle_type_code_2'}, inplace=True)
```

```python
gdf.head()
```

```python
gdf.dtypes
```

```python
gdf.number_of_cyclist_killed = gdf.number_of_cyclist_killed.astype(int)
gdf.number_of_pedestrians_killed = gdf.number_of_pedestrians_killed.astype(int)
gdf.number_of_motorist_killed = gdf.number_of_motorist_killed.astype(int)
```

```python
gdf.number_of_pedestrians_killed.sum()
```

```python
vehicles = pd.DataFrame()

for i in range (1, 6):
       vehicle_code = 'vehicle_type_code_' + str(i)
       contributing_factor  =  'contributing_factor_vehicle_' + str(i)
       vehicles  = vehicles.append(gdf[['number_of_cyclist_killed', 'number_of_motorist_killed',
       'number_of_pedestrians_killed', vehicle_code,
       contributing_factor]].rename(columns = {vehicle_code : 'vehicle_type_code',
       contributing_factor : 'contributing_factor_vehicle'}))
       
vehicles
```

```python
vehicles['vehicle_type_code'] = vehicles['vehicle_type_code'].str.lower()
```

```python
vehicles.vehicle_type_code.unique()
```

```python
mapper  = {   'sedan' : 'small_car', '4 dr sedan' : 'small_car','2 dr sedan' : 'small_car', 'convertible' : 'small_car',
            'pick-up truck' : 'suv/pickup', 'station wagon/sport utility vehicle' : 'suv/pickup', 'sport utility / station wagon' : 'suv/pickup',
            'chassis cab' : 'suv/pickup',
            'bike' : 'bicycle',
            'e-bik': 'e-bike',
            'taxi' : 'tlc_car', 'livery vehicle' : 'tlc_car',
            'e-scooter' : 'motorcycles_scooters', 'motorcycle' : 'motorcycles_scooters',  'e-sco' : 'motorcycles_scooters', 'motorscooter'  : 'motorcycles_scooters',
            'motorbike' : 'motorcycles_scooters', 'minibike' : 'motorcycles_scooters', 'moped' : 'motorcycles_scooters', 'minicycle' : 'motorcycles_scooters',
            'escooter s' : 'motorcycles_scooters', 'motor scoo' : 'motorcycles_scooters', 'e- scooter' : 'motorcycles_scooters',
            'passenger vehicle' : 'van', 'refrigerated van' : 'van',
            'box truck' : 'small_truck', 'small com veh(4 tires) ' : 'small_truck', 'armored truck' : 'small_truck',
            'garbage or refuse' : 'large_truck',
            'tractor truck gasoline' : 'large_truck', 'tow truck / wrecker' : 'large_truck',  'beverage truck' : 'large_truck', 
            'tractor truck diesel'  : 'large_truck', 'flat bed'  : 'large_truck', 'flat rack'  : 'large_truck', 'open body' : 'large_truck', 'carry all' : "large_truck",
            'dump'  : 'large_truck', 'concrete mixer'  : 'large_truck', 'tanker'  : 'large_truck', 'large com veh(6 or more tires)'  : 'large_truck',
            'electric s' : 'unknown', 'pk' : 'unknown',
            'snow plow' : 'construction_utility', 'backhoe lo' : 'construction_utility', 'fork lift' : 'construction_utility'
        }


vehicles['vehicle_type_code'] = vehicles['vehicle_type_code'].map(mapper).fillna(vehicles['vehicle_type_code'])
```

```python
table  = pd.pivot_table(vehicles, values=['number_of_cyclist_killed', 'number_of_motorist_killed', 'number_of_pedestrians_killed'], index=['vehicle_type_code'], aggfunc=np.sum)
```
